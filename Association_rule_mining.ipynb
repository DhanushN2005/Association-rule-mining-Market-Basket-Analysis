```python
#market basket analysis
#1.eda
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
d=pd.read_csv("/content/Market_Basket_Optimisation.csv")
```

```python
d.head()
```

```python
d.isna().sum()
```

```python
transactions = []
for i in range(d.shape[0]):
    transactions.append([str(d.values[i, j]) for j in range(d.shape[1]) if str(d.values[i, j]) != 'nan'])
```

```python
!pip install apyori
```

```python
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2,max_length=3)
results=list(rules)
```

```python
results
```

```python
result=pd.DataFrame(results)
```

```python
result
```

```python
lhs = []
rhs = []
supports = []
confidences = []
lifts = []

for result in results:
    for statistic in result.ordered_statistics:
        lhs.append((statistic.items_base))
        rhs.append((statistic.items_add))
        supports.append(result.support)
        confidences.append(statistic.confidence)
        lifts.append(statistic.lift)

rules_df = pd.DataFrame({
    'lhs': lhs,
    'rhs': rhs,
    'Support': supports,
    'Confidence': confidences,
    'Lift value': lifts
})
rules_df
```
